```python
import os
import analyse
import sys
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
p_directory='../../1Diplom/IOHProfiler/Results/experiment_3.1/'
```

```python
def prepare_res(inds, results, border):
    inds = inds[1:]
    results = results[1:]
    new_inds = []
    new_results = []
    for i, r in zip(inds, results):
        if i > border:
            break
        new_inds.append(i)
        new_results.append(r)
    return new_inds, new_results
```

```python
def get_all_results():
    mega_dict = {}
#     for func_id, best_f in [('1', 100)]:
    for func_id, best_f in [('1', 100), ('2', 100), ('4', 50), ('5', 90), ('6', 33), ('7', 100), ('8', 51), ('9', 100), ('10', 100), ('11', 50), ('12', 90), ('13', 33), ('14', 100), ('15', 51), ('16', 100), ('17', 100)]:
        directory = p_directory + func_id
        ress = {}

        fls = [fl for fl in os.listdir(directory + '/all_zips') if fl.endswith('zip')]
#         print fls
        sys.stdout.write('\n')
        for i, fl in enumerate(fls):
            sys.stdout.write('\r\033[K\033[1F\033[K' + ('%.0f' % (float(i*100) / len(fls))) + '%\tAnalyzing:\t' + fl + '\n')
            sys.stdout.flush()
            ress[fl] = analyse.process_zip(directory + '/all_zips/' + fl, best_f)
        mega_dict[func_id] = ress
    return mega_dict
```

```python
omega_res = get_all_results()
```

```python
def borders(func_id):
    if func_id == '1':
        return 2000
    if func_id == '2':
        return 10000
    return 50000
```

```python
def y_borders(func_id):
    if func_id == '1':
        return 48, 102
    if func_id == '4':
        return 23, 52
    if func_id == '7':
        return 35, 102
    if func_id == '9':
        return 35, 102
    if func_id == '2':
        return -20, 102
    if func_id == '11':
        return -7, 52
    if func_id == '14':
        return  -1, 8
    if func_id == '16':
        return -20, 102
    
    if func_id == '5':
        return 40, 92
    if func_id == '6':
        return 15, 35
    if func_id == '8':
        return 20, 52
    if func_id == '10':
        return -2, 102
    if func_id == '12':
        return -4, 92
    if func_id == '13':
        return -2, 35
    if func_id == '15':
        return  -2, 52
    if func_id == '17':
        return -2, 102
    
    return -2, 102
```

```python
def plot_best_fitness(path, inds, results, freq='t_0', name='dunno', l_b=-2, u_b=102):
    if not os.path.exists(os.path.join(path, 'all_zips', 'comb_graphs')):
        os.mkdir(os.path.join(path, 'all_zips', 'comb_graphs'))
    if not os.path.exists(os.path.join(path, 'all_zips', 'comb_graphs', 'best_fitness')):
        os.mkdir(os.path.join(path, 'all_zips', 'comb_graphs', 'best_fitness'))
    
    fig = plt.figure()
    for i in range(0, len(inds), 2):
        plt.plot(inds[i], results[i], '#1f77b4')
        plt.plot(inds[i+1], results[i+1], '#ff7f0e')
    plt.xlabel('evaluations', fontsize='xx-large')
    plt.ylabel('best f(x) since change', fontsize='xx-large')
    plt.grid(True)
    ax = plt.gca()
    ax.set_ylim(l_b, u_b)
    ax.legend(('$(1+1)EA$','$(1+1)EA_{A,b}$'), loc='lower right')
#     ax.text(0.1, 0.1, freq if freq != 't_0' else 'static', 
#             horizontalalignment='center',
#             verticalalignment='center',
#             transform = ax.transAxes,
#             fontsize='xx-large',
#             bbox={'facecolor':'white', 'edgecolor':'grey', 'alpha':0.8})
#     ax.annotate('static', fontsize='xx-large', xy=(5000, 82), xytext=(0, 90),
#                 arrowprops=dict(facecolor='black', arrowstyle='simple'))
#     ax.annotate('50', fontsize='xx-large', xy=(5000, 50), xytext=(0, 40),
#                 arrowprops=dict(facecolor='black', arrowstyle='simple'))
    ax.annotate('static', fontsize='xx-large', xy=(2300, 60), xytext=(0, 80),
                arrowprops=dict(facecolor='black', arrowstyle='simple'))
    ax.annotate('10', fontsize='xx-large', xy=(1000, 0), xytext=(0, -15),
                arrowprops=dict(facecolor='black', arrowstyle='simple'))
    plt.savefig(os.path.join(path, 'all_zips', 'comb_graphs', 'best_fitness', (name + '.png')), dpi=100)
    plt.savefig(os.path.join('comb_graphs', 'best_fitness', (os.path.basename(path) + '-' + name + '.png')), dpi=100)
    plt.close('all')
```

```python
def plot_changes(path, changes, name='dunno', l_b=-0.1, u_b=1.1):
    if not os.path.exists(os.path.join(path, 'all_zips', 'comb_graphs')):
        os.mkdir(os.path.join(path, 'all_zips', 'comb_graphs'))
    if not os.path.exists(os.path.join(path, 'all_zips', 'comb_graphs', 'changes')):
        os.mkdir(os.path.join(path, 'all_zips', 'comb_graphs', 'changes'))

    fig = plt.figure()
    for i in range(0, len(changes), 2):
        plt.plot(changes[i], '#1f77b4')
        plt.plot(changes[i+1], '#ff7f0e')
    plt.xlabel('number of last change', fontsize='xx-large')
    plt.ylabel('success rate', fontsize='xx-large')
    plt.grid(True)
    ax = plt.gca()
    ax.set_ylim(l_b, u_b)
    ax.legend(('$(1+1)EA$','$(1+1)EA_{A,b}$'), loc='lower right')
    plt.savefig(os.path.join(path, 'all_zips', 'comb_graphs', 'changes', (name + '.png')), dpi=100)
    plt.savefig(os.path.join('comb_graphs', 'changes', (os.path.basename(path) + '-' + name + '.png')), dpi=100)
    plt.close('all')
```

```python
def plot_func(func_id, keys):
    algorithms = ['_stat.zip', '_ab.zip']
    
    f_res = omega_res[func_id]

    name = ''
    for i in range(len(keys)):
        name += keys[i][4:]
        name += '&'
    name = name[:-1]
    
    inds = [f_res[key+suff][0] for key in keys for suff in algorithms]
    results = [f_res[key+suff][1] for key in keys for suff in algorithms]
    changes = [f_res[key+suff][2] for key in keys for suff in algorithms]
    
    prep_inds, prep_res = zip(*[prepare_res(ind, result, borders(func_id)) for ind, result in zip(inds, results)])
    l_b, u_b = y_borders(func_id)
    plot_best_fitness(p_directory+func_id, prep_inds, prep_res, freq=keys[i][11:], name=name, l_b=l_b, u_b=u_b)
    plot_changes(p_directory+func_id, changes, name=name)
```
